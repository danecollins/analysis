```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/Biking.xlsx', header=1, index_col=0, skipfooter=3)
df.head(5)
```

```python
# need to delete the top 3 rows
df = df.iloc[3:]
df.head(2)
```

```python
# need to drop last 2 columns
df.drop(columns=['Unnamed: 12', 'Unnamed: 13'], inplace=True)
df.head(1)
```

```python
# check the types
df.dtypes
```

```python
# Note the type of df.C3 is incorrect
# df.C3.astype('float')
# this fails because there is "29,1" in one cell which should be "29.1"
```

```python
df = df.replace({'29,1': '29.1'})
df['C3'] = df.C3.astype('float')
df.C3.sum()
```

```python
df.dtypes
```

```python
# in Notes, we want to replace NaN with an empty string
df['Notes'] = df['Notes'].replace({np.nan: ''})
```

```python
df.head()
```

```python
# next we need to clean up the index
df.index.value_counts().head(5)
# I cleaned this up manually because I could not think of an automated way to 
# remove duplicates
```

```python
# now we have a clean dataset
df.describe()
```

```python
df.to_parquet('data/cycling.pq')
```
