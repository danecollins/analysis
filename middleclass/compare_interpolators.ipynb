```python
%matplotlib inline
import matplotlib.pyplot as plt
import pylab
pylab.rcParams['figure.figsize'] = (12,8)  # set default plot size
import pandas as pd
import numpy as np
from math import log, e
from numpy import arange
```

```python
from income_processing import *
```

```python
amt = [1, 10, 25, 50, 75, 100, 200]
cnts = [320, 2110, 3060, 3230, 1820, 1190, 1850]
cnts = [
10270,
69470,
273780,
273780,
174560,
135700,
189240,
52140,
]
```

```python
county = County(amt, cnts)
linear = pd.Series([county.lin_pct(x) for x in arange(.1,.9,.1)], name='linear', index=arange(.1,.9,.1))
pareto = pd.Series([county.pareto_pct(x) for x in arange(.1,.9,.1)], name='pareto', index=arange(.1,.9,.1))
df = pd.DataFrame([linear, pareto]).T
df['logl'] = np.log10(df.linear)
df['logp'] = np.log10(df.pareto)
```

```python
plt.plot(df.linear, 'r', df.pareto, 'b')
plt.show()
```

```python
plt.plot(df.logl, 'r', df.logp, 'b')
```
