```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [16,12]
import dtutil.configs as dtc
import glob
```

```python
def read_one_project():
    project_name = '10GHz_LNA\\10GHz_LNA.emp'
    df = None
    fdf = None
    rev = 1
    test_type = 'exampletests'
    fn = '//us-els-argo/AWR/Share/EC folders/Dane/regression_logs/v15pq/test/*{}*.pq'.format(
          test_type)
    print(fn)
    files = glob.glob(fn)
    for f in files:
        tmp_df = pd.read_parquet(f)
        new_df = tmp_df[tmp_df.project == project_name]
        if df is None:
            df = new_df.copy()
            fdf = tmp_df
        else:
            df = pd.concat([df, new_df])
            fdf = pd.concat([fdf, tmp_df])
    df = df.sort_values('build')
    df = df.reset_index()
    return df, fdf
```

```python
df, fdf = read_one_project()
df
```

```python
fig = plt.gcf()
fig.set_size_inches(18,18)
tmp = fdf.project.value_counts()

projects = list(tmp[tmp == 1116].index)
for p in projects:
    pdf_ = fdf[fdf.project == p]
    pdf_ = pdf_.sort_values('build')
    pdf = pdf_.reset_index()
    plt.plot(pdf.elapsed.rolling(window=10).mean());
```
