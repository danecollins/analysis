```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [16,12]
```

```python
df = pd.read_csv('/Users/dane/data/a/perf/linux_test_results.csv')
df.info()
```

```python
from dtutil.sqltools import query_to_dataframe, get_cursor_from_url
```

```python
c = get_cursor_from_url()
```

```python
fig = plt.gcf()
fig.set_size_inches(16,8)

metric_name = 'AxMesh:Time(s)$T'

query = """
select pm.test_case, pt.value from pertest_test as pt
    inner join pertest_metric as pm on pt.metric_id = pm.id
where pm.name = '{}'
and build >= '9662.01'
""".format(metric_name)
db = query_to_dataframe(c, query)
print('{} database records'.format(len(db)))
norm_std = []
for tc, dft in db.groupby('test_case'):
    if dft.value.max() != dft.value.min():
        norm_std.append(dft.value.std() / dft.value.mean())
plt.subplot(121)
plt.title(metric_name)
plt.xlabel('Windows')
plt.hist(norm_std, bins=20, range=(0, 2));

plt.subplot(122)
norm_std = []
for tc, dft in df[df.metric_name == metric_name].groupby('tc'):
    norm_std.append(dft.value.std() / dft.value.mean())
    
plt.hist(norm_std, bins=20, range=(0, 2));
plt.xlabel('linux')
```
